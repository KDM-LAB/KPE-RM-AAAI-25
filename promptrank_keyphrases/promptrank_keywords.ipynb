## Making json file:

```python
def process_str(s):
    s = s.strip()
    s = s.replace("-", " ").replace("'", "").replace("[", "").replace("]", "")
    keyphrases = s.split(", ")

    length_keyphrases = 0
    for idx, keyphrase in enumerate(keyphrases):
        length_keyphrases += len(keyphrase)
        if length_keyphrases >= 900: # for 1000 char limit in db to store keywords
            return ";"(keyphrases[:idx])
    return keyphrases
```

```python
promptrank_keywords = {}
```

```python
for i in range(len(ta)):
    data = ta.iloc[i,[0,1]]
    promptrank_keywords[data[0]] = {"ta_keywords":process_str(data[1]), "pdf_keywords":""}
```

```python
for i in range(len(pdf)):
    data = pdf.iloc[i,[0,1]]
    try:
        promptrank_keywords[data[0]]["pdf_keywords"] = process_str(data[1])
    except:
        promptrank_keywords[data[0]] = {"ta_keywords":"", "pdf_keywords":process_str(data[1])}
```

```python
import json

with open('promptrank_keywords.json', 'w') as fp:
    json.dump(promptrank_keywords, fp)
```
